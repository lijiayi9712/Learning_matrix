```python
import matplotlib.pyplot as plt
import numpy as np
reward1 = np.load('reward_result.npy')
```

```python
z = reward1[:, :, 80, :].mean(axis=2)
p_1 = [round(i, 2) for i in np.arange(0, 1.05, 0.1)]
p_2 = [round(i, 2) for i in np.arange(0, 1.05, 0.1)]
```

```python
import seaborn as sns

def heatmap(z, p_1, p_2, reward_type):
    ax = sns.heatmap(z, linewidth=0.5, cbar_kws={'label': 'average vehicle speed of out edge'})
    ax.set_xticks(np.arange(len(p_1)))
    ax.set_yticks(np.arange(len(p_2)))
    ax.set_xticklabels(p_1)
    ax.set_yticklabels(p_2)
    ax.set(xlabel='p_1', ylabel='p_2')
    plt.setp(ax.get_yticklabels(), rotation=0, ha="right",
             rotation_mode="anchor")
    plt.title("heatmap for reward type{} reward".format(reward_type))
    #plt.show()

heatmap(z, p_1, p_2, 1)
```

```python
z = reward1[:, :, 80, :].mean(axis=2)
heatmap(z, p_1, p_2, 1)
```

```python
reward2 = np.load('reward_result2.npy')
```

```python
z = reward2[:, :, 95, :].mean(axis=2)
heatmap(z, p_1, p_2, 2)
```

```python
z = reward3[:, :, 90, :].mean(axis=2)
heatmap(z, p_1, p_2, 3)
```
